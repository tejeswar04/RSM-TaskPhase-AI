```python
def CostFunc(x,y,w,b):
    c=0
    for i in range(len(x)):
        c=c+((np.dot(w,x[i])+b)-y[i])**2
    c=c/(2*len(x))
    return c
```

```python
def Gradient(x,y,w,b):
    m=len(x)
    n=len(x[0])
    dj_dw=np.zeros(10)
    dj_db=0
    for i in range(m):
        er=((np.dot(w,x[i])+b)-y[i])
        for j in range(n):
            dj_dw[j] = dj_dw[j] + er * x[i, j]
        dj_db=dj_db+er
    w=w-dj_dw*(0.0001)
    b=b-dj_db*(0.0001)
    return w,b
```

```python
epochs=1000
iterations=[]
costs = []
for i in range(epochs):
    w,b=Gradient(x_train,y_train,w,b)
    costs.append(CostFunc(x_train,y_train,w,b))
    iterations.append(i)
```

```python
costs
```

```python
plt.plot(iterations,costs)
plt.show()
```
